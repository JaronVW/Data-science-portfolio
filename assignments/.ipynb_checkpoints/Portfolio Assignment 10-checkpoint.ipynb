```python
import pandas as pd
Pokemon = pd.read_csv('../files/All_Pokemon.csv')
```

# Correlations between various Pok√©mon stats

```python
Pokemon[['Att', 'Def']].corr()
```

```python
Pokemon.plot.scatter(x='Att', y='Def')
```

Attack and defense are neither strongly positively nor negatively correlated. Which is interesting, because you would expect that a Pok√©mon with high attack would more often have low defense to compensate for the high attack, but this is not always the case.

## Correlation between attack and  special attack

```python
Pokemon[['Att', 'Spa']].corr()
```

```python
Pokemon.plot.scatter(x='Att', y='Spa')
```

As expected, most Pok√©mon either have high attack or high special attack, but not both.

## Correlation between HP and defense

```python
Pokemon[['HP', 'Def']].corr()
```

```python
Pokemon.plot.scatter(x='HP', y='Def')
```

The correlation between HP and defense is very low.
