```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

# 1.Numeric data Correlations

```python
penguins[['bill_length_mm', 'bill_depth_mm','flipper_length_mm','body_mass_g']].corr()
```

## 1.1 Strongest positive correlation: body mass and flipper length

```python
penguins.plot.scatter(x='body_mass_g', y='flipper_length_mm')
```

## 1.2 Strongest negative correlation: bill length and bill depth

```python
penguins.plot.scatter(x='bill_length_mm', y='bill_depth_mm')
```

## 1.3 weakest correlation: bill length and body mass

```python
penguins.plot.scatter(x='bill_length_mm', y='body_mass_g')
```
