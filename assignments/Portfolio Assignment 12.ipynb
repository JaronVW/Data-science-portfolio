# Assignment 12: Bivariate Analysis categorical vs numerical:


```python
import pandas as pd
import seaborn as sns
Pokemon = pd.read_csv('../files/All_Pokemon.csv')
```

```python
Pokemon.head()
```

## 1.1 average HP per Pokémon type

```python
Pokemon.groupby('Type 1')['HP'].mean()
```

```python
sns.barplot(x='HP', y='Type 1', data=Pokemon,orient = 'h')
```

There seems to be a pretty big difference in the average HP Per primary Pokémon type. Dragon Pokémon seem to have the biggest confidence interval, this may be due to the rarity of Dragon types.

## 1.2 average HP for legendary Pokémon vs non-legendary Pokémon

```python
Pokemon.groupby('Legendary')['Att'].mean()
```

```python
sns.barplot(y='Att', x='Legendary', data=Pokemon,)
```

There is a pretty big difference in the attack stat between legendary and non-legendary Pokémon.

## 1.3 average HP per generation

```python
Pokemon.groupby('Generation')['HP'].mean()
```

```python
sns.barplot(x='Generation', y='HP', data=Pokemon,)
```

As we can see there is no large increase or decrease in the average HP per generation. Most of the generations have an average HP of around 70. Personally I expected the average HP to increase as the generations went on.
