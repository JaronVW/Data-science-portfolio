```python
import pandas as pd
Pokemon = pd.read_csv('../files/All_Pokemon.csv')
```

```python
Pokemon.head()
```

```python
def create_contingency_table(dataset, column1, column2):
    return dataset.groupby([column1, column2]).size().unstack(column1, fill_value=0)
```

# 1. Contingency table for Type 1 and Legendary status

```python
create_contingency_table(Pokemon, 'Type 1', 'Legendary').plot(kind='bar', stacked=True)
```

# 2. Contingency table for experience type and Legendary status

```python
create_contingency_table(Pokemon, 'Experience type', 'Legendary').plot(kind='bar')
```
