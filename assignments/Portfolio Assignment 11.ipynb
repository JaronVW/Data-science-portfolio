```python
import seaborn as sns
penguins = sns.load_dataset('penguins')
```

```python
penguins.groupby('species').mean()
```

## Mean of all numeric columns grouped by species

```python
sns.barplot(x='species', y='bill_length_mm', data=penguins)
```

Bill length seems to be the most differentiating feature, gentoo and chinstrap have similar bill lengths, but adelie is much shorter on average.

```python
sns.barplot(x='species', y='bill_depth_mm', data=penguins)
```

Adelie has the deepest bill on average, Gentoo has the shallowest, and Chinstrap is in the middle. the confidence intervals of Adelie and Chinstrap seem to overlap, which means that bill depth is not the best feature to distinguish between these two species.

```python
sns.barplot(x='species', y='flipper_length_mm', data=penguins)
```

Flipper length seems to be the most consistent feature between the three species with all the means being within ~30mm of each other.

```python
sns.barplot(x='species', y='body_mass_g', data=penguins)
```

When it comes to body mass Gentoo is the clear outlier, with the other two species being very similar in mass. The confidence intervals of Adelie and Chinstrap heavily overlap, which means that body mass may not the best feature to distinguish between these two species.

In conclusion, it seems that Gentoo penguins are the biggest, since their body mass and flipper length are the largest.
