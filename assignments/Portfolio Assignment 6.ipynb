```python
Pokemon.head()
```

## 1. Number of Pokémon type occurrences

```python
pokemon_with_one_type = Pokemon[Pokemon['Type 2'].isnull()]
pokemon_with_two_types = Pokemon[Pokemon['Type 2'].notnull()]

two_type_value_counts = pokemon_with_two_types['Type 2'].value_counts() + pokemon_with_two_types['Type 1'].value_counts()
total = pokemon_with_one_type['Type 1'].value_counts() + two_type_value_counts

total.plot.barh()
```

In the graph above, column type 1 and type 2 are combined to show the total number of occurrences of each type. Since Pokémon can have two types simultaneously we can get a more complete result combining them

## 2. Average Pokemon HP

```python
Pokemon["HP"].median()
```

```python
Pokemon["HP"].mean()
```

```python
Pokemon["HP"].plot.box()
```

This Boxplot shows Pokémon HP, we can see that the average Pokémon HP lies just above 50. The max value excluding is around ~140 but there are a lot of outliers. Pokémon can stats can vary greatly so this is a expected result.
